```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('pemandangan.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Matriks dan vektor
print(gray[:5, :5])
print(gray.flatten()[:10])

# Parameter
h, w = gray.shape
bit_depth = 8

size_bytes = w * h * bit_depth / 8

print("Resolusi:", w, "x", h)
print("Ukuran:", size_bytes/1024, "KB")

# Cropping
crop = gray[100:400, 100:400]

# Resize
resize = cv2.resize(gray, (300, 300))

# Rotasi
center = (w//2, h//2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotate = cv2.warpAffine(gray, matrix, (w, h))

# Flip
flip = cv2.flip(gray, 1)

# Tampilkan
titles = ["Asli", "Gray", "Crop", "Resize", "Rotate", "Flip"]
images = [img_rgb, gray, crop, resize, rotate, flip]

plt.figure(figsize=(12,8))

for i in range(len(images)):
    plt.subplot(2,3,i+1)

    if len(images[i].shape) == 2:
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(images[i])

    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
